```python
y = df['y'] # Target vector
X =  df.drop(columns=['y'])#Design Matrix
X
```

```python
##Using Cross validation to test whats the score with all features included 
maxScore = 0.0
knn = pa.knnClassifier()
model_args = {'n_neighbours' : 1,'weights':'uniform','distance' :'Euclidean'}
result = pa.sFold(5,X,y,knn,pa.compute_F1_score,**model_args )
maxScore = result['Average error']
print(maxScore)
```

```python
## Use backward search(Warping feature selection) to test all possible feature combination
## to run KNN for 2^11 times, as each run takes few seconds
#y = np.array(y)
for j in range(len(X.columns)): 
    index = -1
    for i,cols in enumerate(X.columns):
        fs_df = X    
        fs_df = fs_df.drop(columns = fs_df.columns[i])
        #fs_df = np.array(fs_df)
        knn = pa.knnClassifier()   
        
        model_args = {'n_neighbours' : 1,'weights':'uniform','distance' :'Euclidean'}
        result = pa.sFold(5,fs_df,y,knn,pa.compute_F1_score,**model_args )
        validation_accuracy = result['Average error']  
        
        if validation_accuracy > maxScore:
            maxScore = validation_accuracy
            index = i
    if(index != -1):
        print('This is Dropped:',X.columns[index])
        X = X.drop(columns = X.columns[index])
    print(X.columns) 
    print('\n','='*30)
```

```python
best_features_accuracy_f1 = ['volatile acidity', 'citric acid', 'residual sugar', 'pH', 'sulphates',
       'alcohol']
best_features_accuracy = ['volatile acidity', 'density', 'pH', 'sulphates', 'alcohol']
best_df = df[best_features_accuracy]
```
