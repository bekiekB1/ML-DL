```python
#In case of NA

#just drop the NA
df.dropna()

#to fill Na with 0
df["G3"] = df["G3"].fillna(0)

#to fill NA with median values
median = df["G3"].median() 
df["G3"] = df["G3"].fillna(median, inplace=True)

# Using imputer of sklearn
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
df_numeric = df.select_dtypes(include=numerics)

imputer.fit(df_numeric)
df["G3"].value_counts()
```

```python
# Write a function to count 0s in a given column

count = 0

def zeroCount(dataFrame, columnName):
    noOfZeros = 0
    
    for i in range(dataFrame.shape[0]):
        if (dataFrame[columnName].get(i) != None
           ) and dataFrame[columnName].get(i) <= 10:
            
            noOfZeros = noOfZeros + 1
    return noOfZeros

count = zeroCount(df, ' shares')

print("Number of shares less than 10 in the Target Column: %d" % count)
```
